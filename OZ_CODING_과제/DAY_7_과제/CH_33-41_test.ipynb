[문제 1]

사용자로부터 두 개의 숫자를 입력받아 나눗셈을 수행하는 프로그램을 작성하세요.

[예외처리]
- 입력 값이 숫자가 아닐 때 발생할 수 있는 ValueError
- 나누는 수가 0일 때 발생할 수 있는 ZeroDivisionError
- 위 2가지 예외처리 후 사용자에게 이해하기 쉬운 메시지 작성 후 출력되도록 코드를 작성

````python
try:
    # 사용자로 부터 두가지의 숫자를 입력받기
    a, b = map(int, input('두개의 숫자를 입력하세요(숫자는 반드시 ,로 구분할 것) : ').split(","))
    # 사용자에게 받은 값을 정석으로 푸는 방법
    user_number = a / b # 결과를 정수로 받기 위한 예) a // b = 2
# 사용자가 정수(int)를 제외한 값을 입력한 경우의 에러코드
#    이유는 예외가 발생할 수 있는 코드에 map(int, input())와 값을 계산하는 것을
#    넣었기에 이곳에는 당연히 정수만이 와야한다.
# 만약 구하는 값이 이름과 나이 처럼 구분지어지지 않은 값이라면
# 작성자가 직접 if not isinstance로 구분지어줘야한다.
except ValueError as exception:
    print("유효한 숫자가 아닙니다. 예) 1, 2")
    print("Exception:", exception)
# 사용자가 0으로 나누려할 경우의 에러코드 예) 10 / 0
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
# 정상적으로 작성시 작동 결과
else:
    print(f'입력한 값은 {user_number}입니다.')
# 마무리 멘트
finally:
    print("프로그램을 종료합니다.")

# 위에는 너무 복잡한 구조의 작성법
````

[문제 3]

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요.

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
numbers = []

# while 반복문을 만든다.
# done을 입력하 시 종료됀다고 안내해준다
while True:
    user = input("숫자를 입력하세요. 종료 시 (done)입력")
#   사용자에게 done을 입력 받으면 break 멈추어준다.
    if user == "done":
        break
#   사용자가 입력한 값을 정수로 받아와 number에 넣어준다.
#   number가 받은 정수를 numbers의 리스트 안에 넣는다.
    try:
        number = int(user)
        numbers.append(number)
#   발생할 수 있는 예외처리를 알려준다.
    except ValueError:
            print("숫자를 입력하세요.")

print("입력된 숫자 :", numbers)
````

````python
class NegativeNumberError(Exception):
    pass

try:
    a = int(input("숫자를 입력해주세요."))

    if 0 > a:
        raise NegativeNumberError
    print(f"입력한 숫자는 {a}입니다.")

except NegativeNumberError:
    print("양수를 입력해주세요.")
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요.

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def timer_decorater(x):
    def wrapper():
        start = time.time()
        x()
        end = time.time()
        print(f'함수 실행 시간:{start - end:.2f}초')
    return wrapper

@timer_decorater
def test_func():
    time.sleep(0)

test_func()
````

[문제 8]

여러분이 자주 사용하는 수학적 연산<br>
(예: 삼각형 넓이, 원의 넓이, 직육면체의 넓이)을 수행하는 함수들을 포함하는 mymath.py라는 이름의 모듈을 만드세요.

이 모듈을 사용하여 작성된 함수들을 실제로 임포트하고 사용하는 간단한 프로그램을 작성하세요.

````python
### mymath.py 파일 안에 모듈 생성

import math

def triangle(base, heigh):
    return base * heigh / 2

# 파이 상수 가져와서 사용
def circle(radius):
    return math.pi * radius ** 2

def prism(width, height, depth):
    return 2 * (width * height + width * depth + height * depth)

### mymath.py에 있는 모듈을 가져와서 사용.

import mymath

print("삼각형의 넓이 :", mymath.triangle(10, 5))
print("원의 넓이 :", mymath.circle(5))
print("직육면체의 넓이 :", mymath.prism(2, 3, 4))
````
