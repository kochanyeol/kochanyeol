__[문제0]__


오즈 코딩 스쿨 백엔드 8기 과정에 참여한 수강생 중 3명의 수강생이 자율적으로 스터디 그룹을 결성했습니다.

스터디 그룹 내에는 한 주간 배운 내용으로 매주 쪽지 시험을 보는 룰이 있습니다. 점수는 채점자 마음입니다.

쪽지 시험의 점수를 직접 입력받고, 점수의 합을 출력해주세요!


<br>



````python
# tuple(불변) = 학생 수 변동 없음
oz_cod_8 = ("student_1", "student_2", "student_3")
# input으로 입력받고, 문자열을 map을 묶어서 int로 형변환,
# 하나의 list로 담아두기
scores = list(
    map(
        int,
        input("띄어쓰기를 사용하여 점수 3개를 입력하시오: ").split())
)
# zip을 이용해 oz_cod_8과 socres의 같은줄을 묶고
# dict으로 이름표 새기기
student_scores = dict(zip(oz_cod_8, scores))
# sum을 사용해 모든 점수 합산
total_scores = sum(scores)

print(f'학생별 점수 : {student_scores}')
print(f'점수 총합 : {total_scores}')
````

__[문제3]<br>__

다음 리스트에서 최댓값과 최솟값, 그리고 평균을 출력해주세요.
<br>

```
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
```



````python
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]

high_nums = max(nums)
low_nums = min(nums)
avg_nums = int(sum(nums) / len(nums))

# 오름차순 정렬 상태에 사용가능
# print(f'최댓값 : {nums[-1]}')
# print(f'최소값 : {nums[0]}')

print(f'최댓값 : {high_nums}')
print(f'최소값 : {low_nums}')
print(f'평균값 : {avg_nums}')
````
