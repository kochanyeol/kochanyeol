[문제 4] 사용자의 이름, 나이, 성별을 출력하는 함수 print_info를 작성하세요. 이 때, 모든 매개변수는 키워드 매개변수로만 받도록 하세요.

```python
def print_info(*, name : str, age : int, sex : str) -> None:
    """
    사용자의 이름과 성별은 문자열로
    나이는 정수로 기입합니다.
    """
    if not (
        isinstance(name, str)
        and isinstance(sex, str)
        and isinstance(age, int)
    ):
        raise TypeError('이름과 성별은 문자열, 나이는 정수로 입력해주세요.')

    print(f'이름 : {name}, 나이 : {age}, 성별 : {sex}')

print_info(name="고찬열", age=29, sex="남성")
```

[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n:int):
    """
    기입하는 n의 숫자는
    정수로 기입한다.
    """
    total_numbers = []

    for x in range(0, n+1, 2):
        total_numbers.append(x)
    return total_numbers

result = odd_numbers(20)
print(result)
```

[문제 8] 사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max() -> tuple[int, int]:
    a = int(input("숫자를 입력하세요.: "))
    b = int(input("숫자를 입력하세요.: "))
    c = int(input("숫자를 입력하세요.: "))

    result = (a,b,c)

    return max(result), min(result)

low, high = min_max()
print(f'최대값 : {high}')
print(f'최소값 : {low}')
```

[문제 9] 주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(a : int) -> bool:
    """
    짝수가 맞을 경우
    조기 리턴하여 True를 알려준다.
    """
    if a % 2 == 0:
        return True
    return False

default = is_even(13)
print(default)
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):
    """
    자신보다 낮은 1까지의 모든
    수를 곱한다.
    """
    if not isinstance(n, int):
        raise TypeError('정수만 입력할 수 있습니다.')
    if n < 0:
        raise TypeError('0 이상의 정수만 입력할 수 있습니다.')
    if n <= 1:
        return 1
    else:
        return n * factorial(n-1)

result = factorial(5)
print(result)
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(word : list[str]) -> None:
    for a in word:
        print(len(a))

words = ["방가", "gasga", "dsfsadf"]
print_lengths(words)
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(a:str, b:str) -> str:
    """
    위에 사용한 조기 리턴과 비교 연산을 사용하여
    a가 b보다 크거나 작으면 리턴하여 a를 반환한다.
    """
    if not isinstance(a, str) or not isinstance(b, str):
        raise TypeError("문자열을 입력해주세요.")
    if len(a) >= len(b):
        return a
    return b

print(longer_string("방가", "graega"))
```
